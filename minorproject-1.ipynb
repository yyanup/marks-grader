```python
#Problem Statement Develop a program that analyzes student marks using Python and NumPy
#logic :  set the subject list and no. of subjects and a marks dictionary --->Ask for marks input subject wise using for loop ----> make another variable to convert the input into numarray ---->make functions for  ○ Total marks ○ Average marks ○ Highest & lowest marks ----->ddd
#make a grading function using if/ elif statements -----> print command for the variables ----> calling the grading function

import numpy as np

# Define your standard subjects here
subject_names = ["English", "Science", "Math", "History", "Art"]
num_subjects = len(subject_names)

marks_dict = {}
for subject in subject_names:
    mark = float(input(f"Enter marks for {subject}: "))
    marks_dict[subject] = mark

marks_array = np.array(list(marks_dict.values()))
total_marks = np.sum(marks_array)
average_marks = np.mean(marks_array)
highest_mark = np.max(marks_array)
lowest_mark = np.min(marks_array)

def grade_student(average_marks):
    if average_marks >= 85:
        return "A"
    elif average_marks >= 70 and average_marks < 85:
        return "B"
    elif average_marks >= 50 and average_marks < 70:
        return "C"
    else:
        return "Fail"

print("Marks entered:", marks_dict)
print("Marks array:", marks_array)
print("Total Marks:", total_marks)
print("Average Marks:", average_marks)
print("Highest Mark:", highest_mark)
print("Lowest Mark:", lowest_mark)
grade = grade_student(average_marks)
print("Grade: ", grade)
```
